```
%sql
select count(*)
from orders
group by ShipCountry
order by count(1) desc
```

```
%sql

select p.ProductID, p.ProductName, s.CompanyName
from products as p
join suppliers as s
on p.SupplierID=s.SupplierID
```

```
%sql

select ShipCountry, avg(Freight)
from orders
group by ShipCountry
order by avg(Freight) desc
```

```
%sql

select c.CompanyName, c.ContactName, c.Phone, count(*) as numOrders
from orders as o
join customers as c
on o.CustomerID = c.CustomerID
group by c.CompanyName, c.ContactName, c.Phone
order by numOrders desc
limit 10
```

```
%sql
select c.CompanyName, sum(od.UnitPrice*od.Quantity) as total_spending
from orders as o
join order_details as od
on o.OrderID=od.OrderID
join customers as c
on c.CustomerID=o.CustomerID
group by c.CustomerID, c.CompanyName
order by total_spending desc
limit 10
```
